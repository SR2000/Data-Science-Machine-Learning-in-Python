```python
df_ds = pd.read_csv('https://cocl.us/datascience_survey_data',index_col= 0)
df_ds
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_ds.sort_values(by= 'Very interested', ascending= False, inplace= True)
df_ds
```

```python
#df_ds= df_ds.apply(lambda x: x/x.sum(), axis= 1)
df_ds= df_ds.apply(lambda x: (x/2233)*100, axis= 1)
df_ds= df_ds.apply(lambda x: round(x,2), axis= 1)
#df_ds= df_ds.apply(lambda x: round(x,2), axis= 1)

df_ds
```

```python
ax = df_ds.plot(kind='bar', width= 0.8, figsize=(20, 8), color=['#5cb85c', '#5bc0de', '#d9534f'])
ax = plt.gca()
ax.set_title('Percentage of Respondents\' Interest in Data Science Areas').set_fontsize('16')
ax.legend(fontsize='14')

for label in ax.xaxis.get_ticklabels():
    label.set_fontsize(14)
    
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.axes.get_yaxis().set_ticks([])

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.4}'.format(height), (x+0.01, y + height + 0.6), fontsize= 14)

plt.show()
```
